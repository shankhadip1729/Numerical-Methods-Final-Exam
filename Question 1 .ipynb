```python
import numpy as np
from scipy.integrate import simps
import matplotlib.pyplot as plt
def ro(r):
    return(3*((1+r**2)**(-5/2))/(4*np.pi))
def f(r):
    return(4*np.pi*(r**2)*ro(r))

def mass(a):
    r_list = np.linspace(0,a,1000)
    return(simps (f(r_list),x = r_list))

r_list2 = np.linspace(0,10,1000)
print(mass(10))

plt.plot(r_list2,mass(r_list2),label = 'plot of mass vs radius')
plt.title('mass vs radius')
plt.xlabel('radius(m)')
plt.ylabel('mass(Kg)')
plt.legend()
plt.show

```
