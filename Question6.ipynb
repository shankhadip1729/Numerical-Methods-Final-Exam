```python
import numpy as np
from scipy.integrate import simps
from scipy.optimize import newton
import matplotlib.pyplot as plt

def f(x):
    return(-1/(2.5*x))
# Finding t as function of n
def t(n):
    x=np.linspace(40,n,1000)
    return(simps(f(x),x=x))

# Finding the value of n when t = 0
def g(x):
    return(t(x)-10)
root = newton(g,0.001)

# Plotting n for t = 0 to 10 s
n_list = np.linspace(40,root,1000)
plt.plot(t(n_list),n_list)
plt.xlabel('t(in s)')
plt.ylabel('n(in cm^(-3))')
plt.show()
   
```
